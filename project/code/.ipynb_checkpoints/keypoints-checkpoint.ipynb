### Check overlap

```python
keypoints_dict = {}

for filename in keypoints_filenames:
    prefix, name, params = parse_filename(filename)
#     cov_filename = give_filename(name, prefix='covariance', params=params, extension='npy')

    cloud_ply = read_ply(os.path.join(saved_data_path, filename))
    keypoints = cloud_ply['keypoints']
    if 'factor' not in name:
        keypoints_dict[name] = (np.where(keypoints==1)[0]/5).astype('int')
    else:
        keypoints_dict[name] = np.where(keypoints==1)[0]
```

```python
[len(v) for k,v in keypoints_dict.items()]
```

```python
def jaccard_similarity(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
```

```python
keys = sorted(list(keypoints_dict.keys()))
jaccards = np.zeros((len(keys), len(keys)))

for i,k1 in enumerate(keys):
    for j,k2 in enumerate(keys):
        jaccards[i,j] = jaccard_similarity(keypoints_dict[k1], keypoints_dict[k2])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(jaccards, vmin=0, vmax=1, yticklabels=keys,xticklabels=keys)
plt.xticks(rotation=70) 

plt.tight_layout()
plt.savefig('../images/overlaps.png')
```
